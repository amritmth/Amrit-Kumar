### Bar plot

```python
activity=["Sleep", "Reading", "Office","Home Chores", "Others"]
hours=[4, 6, 8,3,3]
plt.bar(activity, hours, color="g")
plt.show()
```

### Horizanatal  Bar Plot

```python
activity=["Sleep", "Reading", "Office","Home Chores", "Others"]
hours=[4, 6, 8,3,3]
plt.barh(activity, hours, color="g")
plt.show()
```

### Stacked Plot

```python
activity=["Sleep", "Reading", "Office","Home Chores", "Others"]
day1=[4, 6, 8,3,3]
day2=[5, 6, 8,2,3]
day3=[10, 4, 2,4,4 ]
plt.stackplot(activity, day1, day2, day3, colors=["r", "m", "c"])
plt.show()
```
