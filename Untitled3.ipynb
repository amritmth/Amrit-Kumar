```python
a1=np.linspace(0,2*np.pi, 400)
b1=np.sin(a1**2)
plt.plot(a1,b1)
```

```python
fig, ((ax0,ax1), (ax2,ax3))=plt.subplots(2,2, sharex=True, sharey=True)
ax0.plot(a1,b1,color="red")
ax1.plot(a1,b1, color="blue")
ax2.plot(a1,b1, color="orange")
ax3.plot(a1,b1, color="cyan")
```

```python


fig, ((ax0,ax1), (ax2,ax3))=plt.subplots(2,2, sharex=True, sharey=True, 
                                         subplot_kw=dict(projection="polar"))
fig.patch.set_facecolor("black")

ax0.plot(a1,b1,color="red")
ax1.plot(a1,b1, color="blue")
ax2.plot(a1,b1, color="orange")
ax3.plot(a1,b1, color="cyan")
plt.show() 
```

#### Bar Graph

```python
product=["orange", "Apple", "Mango", "Banana"]
sale=[100, 160,120,210]
```

```python
plt.bar(product, sale, color=["r","b","k","c"])
plt.show()
```

```python
plt.barh(product, sale, color=["r","b","k","c"])
plt.show()
```

### Stacked Bar Graph 

```python
Profit=[40, 35, 25,70]
```

```python

plt.bar(product, sale, color="b")
plt.bar(product, Profit, bottom=sale, color="red")
plt.legend(["Sale", "Profit"])
plt.show()
```

### Pie Graph 

```python
plt.pie(x=sale, labels=product)
plt.show()
```

```python
plt.pie(x=sale, labels=product, shadow=True, explode=[0,0,.2,0], 
        autopct="%1.1f%%",labeldistance=1.2 , pctdistance=.8, startangle=90)
plt.show()
```
