#### creating columns based on condition

```python
churn["Left_yes_no"]=np.where(churn["Exited"]==1, "Left", "Not Left")
```

```python
churn.head()
```

```python
churn["Class_money"]=np.where(churn["Balance"]> 10000, "Rich", "Average")
```

```python
churn.head()
```

```python
def myfun(x):
    if x<=5000:
        return "Below Average"
    elif x >5000 and x <= 10000:
        return "Average"
    elif x >10000 and x<=50000:
        return "Above Average"
    else:
        return "Rich"
```

```python
churn["Balance_cat"]=churn["Balance"].apply(myfun)
```

```python
churn.head(1)
```

```python
churn["Balance_cat"].value_counts()
```

```python
def cibil1(x):
    if x>=300 and x<550:
        return "poor"
    elif x>=550 and x<650:
        return "Average"
    elif x>=650 and x<750:
        return "Good"
    elif x>=750 and x<=900:
        return "Excellent"
    else:
        return "None"  
```

```python
churn["CreditScore_Cat"]=churn["CreditScore"].apply(cibil1)
```

```python
churn["CreditScore_Cat"].value_counts()
```
